```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
import stan_jupyter as stan
import pandas as pd
```

```python
size=100
price = []
competitor_price = []
account = []
sku = []

alpha = 50
beta = -.3
gamma = .17
sigma = 10

for i in [.85, .95, 1, 1.05, 1.15]: # accounts
    for j in [.75, .8, 1, 1.2, 1.3]:
        price.extend(stats.norm.rvs(loc=100*i*j, scale=11*i*j, size=size).tolist())


for i in [.8, .9, 1, 1.1, 1.2]: # accounts
    for j in [.5, .75, 1, 1.25, 1.5]:
        competitor_price.extend(stats.norm.rvs(loc=100*i*j, scale=11*i*j, size=size).tolist())
        

for i in [.8, .9, 1, 1.1, 1.2]: # accounts
    for j in [.5, .75, 1, 1.25, 1.5]:
        account.extend([i]*size)
        sku.extend([j]*size)
```

```python
df = pd.DataFrame({'account':account,
                   'sku':sku,
                   'price':price,
                   'competitor_price':competitor_price,
                   'alpha':alpha,
                   'beta':beta,
                   'gamma':gamma})

df['sigma'] = np.random.normal(1,.15, len(df)) # to add some noise
df['quantity'] = ((df.alpha + df.price*df.beta + df.competitor_price*df.gamma)*df.sigma).astype(int)

df.account = df.groupby('account').ngroup()+1
df.sku = df.groupby('sku').ngroup()+1
df = df[["sku", "account", "price", "competitor_price", "quantity"]]
```

```python
sns.relplot(x="price", y="quantity", data=df, hue="sku", height=10, aspect=1.5)
```

```python
df.to_csv("sim_data.csv", index=False)
```
