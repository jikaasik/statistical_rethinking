```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import arviz as az

import stan_jupyter as stan
import pandas as pd
```

## Question 1

```python
df = pd.read_csv("../data/foxes.csv", delimiter=";")
```

```python
model_data = {"N":len(df),
              "area":df.area.tolist(),
              "food":df.avgfood.tolist()}

with open("models/w3_1.stan") as f:
    model_code = f.read()
```

```python
%%capture
posterior = stan.build(model_code, model_data)
```

```python
%%capture
fit = posterior.sample(num_chains=8, num_samples=1000)
```

```python
az.summary(fit)
```

```python
az.plot_trace(fit)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x=df.area-df.area.mean(), y=df.avgfood, alpha=.2)
for i in range(1, 100):
    plt.plot(
        df.area-df.area.mean(),
        fit['alpha'][0][i]+fit['beta'][0][i]*(df.area-df.area.mean()),
        color='black',
        alpha=.05)
plt.title("Avg food ~ area")
plt.xlabel("Area (deviation from mean)")
plt.ylabel("Avg food")
```

Larger areas are associated with considerably larger amounts of food, on average.

## Question 2

### Total effect of food on the weight of foxes

In this case, we need to exclude group size `G`. Including it would shut that route and attribute to `G` the part of `F` 's effect that is mediated by `G`. In other words, we want to just consider the model $weight \sim \alpha + food*\beta$.

Because the scale of food is roughly 0 to 1, let's standardize it to make interpretation more reasonable. Without standardization, the effect would be massive and hard to interpred intuitively, because it would represent a move from no food to nearly the maximum possible food.

```python
df["std_food"] = (df.avgfood - df.avgfood.mean()) / df.avgfood.std()

model_data = {"N":len(df),
              "weight":df.weight.tolist(),
              "food":df.std_food.tolist()}

with open("models/w3_2.stan") as f:
    model_code = f.read()
```

```python
%%capture
posterior = stan.build(model_code, model_data)
```

```python
%%capture
fit = posterior.sample(num_chains=8, num_samples=1000)
```

```python
az.plot_trace(fit)
```

```python
az.summary(fit)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x=df.std_food-df.std_food.mean(), y=df.weight, alpha=.2)
for i in range(1, 100):
    plt.plot(
        df.std_food-df.std_food.mean(),
        fit['alpha'][0][i]+fit['beta'][0][i]*(df.std_food-df.std_food.mean()),
        color='black',
        alpha=.05)
plt.title("Weight ~ Food")
plt.xlabel("Food (deviation from mean)")
plt.ylabel("Avg weight")
```

### Direct effect of food on weight

To get at the direct effect, we need to remove the effect that is mediated by group size `G`. Our new model is $weight \sim \alpha + food*\beta + group size*\gamma$

```python
df["std_groupsize"] = (df.groupsize - df.groupsize.mean()) / df.groupsize.std()

model_data = {"N":len(df),
              "weight":df.weight.tolist(),
              "food":df.std_food.tolist(),
              "group_size":df.std_groupsize.tolist()}

with open("models/w3_2b.stan") as f:
    model_code = f.read()
```

```python
%%capture
posterior = stan.build(model_code, model_data)
```

```python
%%capture
fit = posterior.sample(num_chains=8, num_samples=1000)
```

```python
az.summary(fit)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x=df.std_food-df.std_food.mean(), y=df.weight, alpha=.2)
for i in range(1, 100):
    plt.plot(
        df.std_food-df.std_food.mean(),
        fit['alpha'][0][i]+fit['beta'][0][i]*(df.std_food-df.std_food.mean()),
        color='black',
        alpha=.05)
plt.title("Weight ~ Food, stratified by group size")
plt.xlabel("Food (deviation from mean), stratified by group size")
plt.ylabel("Avg weight")
```

In this case, we need to include `X`, `S`, and `A`. If `S` and `A` only affected `Y` through `X`, we could exclude them and the causal effect could still validly be attributed to `X`. However, because they also directly affect `Y`, excluding them will result in misattributing some of their effect to `X`. By including all three, we solve that problem, but only the effect of `X` remains causal, because the effects of `S` and `A` are being partially muted by the inclusion of `X`.
