```python
import numpy as np
import matplotlib.pyplot as plt
import arviz as az
import scipy.stats as stats
```

## Question 1

```python
size = 1000
p_grid = np.linspace(0, 1, size)
prior = np.repeat(1, size)

likelihood = stats.binom.pmf(4, 11, p_grid)
posterior = likelihood * prior
posterior = posterior / sum(posterior)
```

```python
samples = np.random.choice(p_grid, p=posterior, size=size, replace=True)
```

```python
np.mean(samples)
```

```python
az.plot_kde(samples)
```

## Question 2

```python
prior = np.where(p_grid<.5, 0, 1)

likelihood = stats.binom.pmf(4, 6, p_grid)
posterior = likelihood * prior
posterior = posterior / sum(posterior)

samples = np.random.choice(p_grid, p=posterior, size=size, replace=True)
```
