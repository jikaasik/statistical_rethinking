## Question 4

__Part 1__

```python
def posterior_grid_approximate(grid_points=10000, water=7, tosses=10, bias=0):
    
    p_grid = np.linspace(0, 1, grid_points)
    prior = np.repeat(1, grid_points)
    
    biased_water = stats.binom.rvs(n=water, p=1-bias, size=10000)
    likelihood = stats.binom.pmf(biased_water, tosses, p_grid)
    
    unstd_posterior = likelihood * prior
    
    posterior = unstd_posterior / unstd_posterior.sum()
    
    return p_grid, posterior
```

```python
p_grid, posterior = posterior_grid_approximate(bias=.2)
biased_samples = np.random.choice(p_grid, p=posterior, size=10000, replace=True)

p_grid, posterior = posterior_grid_approximate()
unbiased_samples = np.random.choice(p_grid, p=posterior, size=10000, replace=True)
```

```python
print(f"The unbiased proportion is approximately .7, but the biased sampling produces an estimate of ~{round(np.mean(biased_samples), 2)}.")
```

```python
tosses = 20
water = int(.7*tosses)
p_grid, posterior = posterior_grid_approximate(water=water, tosses=tosses)
unbiased_samples = np.random.choice(p_grid, p=posterior, size=10000, replace=True)
```

```python
az.plot_kde(biased_samples, plot_kwargs={"color":"r", "alpha":.7})
az.plot_kde(unbiased_samples)
plt.title("Biased (red) vs. unbiased (blue) parameter estimates")
```

```python
print(f"The unbiased proportion of water is approximately {round(np.mean(unbiased_samples), 2)}.")
```
