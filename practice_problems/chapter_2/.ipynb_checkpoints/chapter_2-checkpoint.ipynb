__2M1__

```python
def posterior_grid_approximate(grid_points=100, water=6, tosses=9):
    
    p_grid = np.linspace(0, 1, grid_points)
    prior = np.repeat(1, grid_points)
    
    likelihood = stats.binom.pmf(water, tosses, p_grid)
    
    unstd_posterior = likelihood * prior
    
    posterior = unstd_posterior / unstd_posterior.sum()
    
    return p_grid, posterior
```

```python
_, ax = plt.subplots(1, 3, figsize=(20,5))

water_rates = ((3, 3), (3,4), (5, 7))

for idx, rate in enumerate(water_rates):
    p_grid, posterior = posterior_grid_approximate(water=rate[0], tosses=rate[1])
    ax[idx].plot(p_grid, posterior)
    ax[idx].set_title(f"{rate[0]}/{rate[1]} tosses result in water")
    ax[idx].set_ylabel("Posterior probability")
    ax[idx].set_xlabel("Pr(Water)")
```

```python
def posterior_grid_approximate(grid_points=100, water=6, tosses=9):
    
    p_grid = np.linspace(0, 1, grid_points)
    prior = np.where(p_grid<.5, 0, 1)
    
    likelihood = stats.binom.pmf(water, tosses, p_grid)
    
    unstd_posterior = likelihood * prior
    
    posterior = unstd_posterior / unstd_posterior.sum()
    
    return p_grid, posterior

_, ax = plt.subplots(1, 3, figsize=(20,5))

water_rates = ((3, 3), (3,4), (5, 7))

for idx, rate in enumerate(water_rates):
    p_grid, posterior = posterior_grid_approximate(water=rate[0], tosses=rate[1])
    ax[idx].plot(p_grid, posterior)
    ax[idx].set_title(f"{rate[0]}/{rate[1]} tosses result in water")
    ax[idx].set_ylabel("Posterior probability")
    ax[idx].set_xlabel("Pr(Water)")
```

__2M3__

$$Pr(earth|land) = \frac{Pr(land|earth) * Pr(earth)}{Pr(land)}$$

- The probability of land = 1.3/2 (1 for Mars, .3 for earth, divided by 2 planets)
- The probability of earth = .5
- The probability of land, given earth = .3

$$Pr(earth|land) = \frac{.3 * .5}{(1+.3) / 2}$$

$$Pr(earth|land) \approx .23$$


```python
(.3*.5) / ((1+.3) / 2)
```

_Alternative_

The second card side being white can happen 2 ways with the white/white card and 1 way with the white/black card (3 ways total). If we remove that 1/3 from the white/black card's probability of having been the first card (now prob weights = 1, 1/3, 1), the new normalizing constant (denominator) becomes:

$$ 1*0 + (1-1/3)*1 + 1*2 = 2\frac{2}{3} $$

If we then divide the 2 ways a black/black card can produce a black side by this normalizing constant, we get 

$$ \frac{2} {2\frac{2}{3}} \approx .75 $$

```python
2/((1-1/3)*1 + 1*2)
```
